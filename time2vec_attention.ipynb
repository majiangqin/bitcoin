```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
from dataclasses import dataclass
from typing import List, Dict, Tuple
import time
import random

@dataclass
class ModelConfig:
    """Optimized configuration for LSTM model"""
    time2vec_dim: int = 32#32
    hidden_dim: int = 128#128  # Reduced to prevent overfitting
    batch_size: int = 32   # Reduced to help with stability
    learning_rate: float = 0.01#5e-5
    num_epochs: int = 100
    max_encoder_length: int = 576 #576  # 4 days of history
    max_prediction_length: int = 144
    clip_grad_norm: float = 1.5 #1.5
    warmup_epochs: int = 3
    patience: int = 10
    min_delta: float = 1e-4
    dropout_rate: float = 0.02# 0.02  
    num_lstm_layers: int = 3 #3  # Reduced number of layers
    num_attention_heads: int = 4

class EarlyStopping:
    """Enhanced early stopping handler"""
    def __init__(self, patience: int = 15, min_delta: float = 1e-5, verbose: bool = True):
        self.patience = patience
        self.min_delta = min_delta
        self.verbose = verbose
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
        self.val_loss_min = float('inf')
        self.val_losses = []  # Track loss history

    def __call__(self, val_loss: float, model: nn.Module) -> bool:
        self.val_losses.append(val_loss)
        
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.verbose:
                print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0
        return self.early_stop

class Time2Vec(nn.Module):
    """Enhanced Time2Vec with periodic embeddings"""
    def __init__(self, input_dim: int, output_dim: int):
        super(Time2Vec, self).__init__()
        self.w0 = nn.Parameter(torch.randn(input_dim, 1))
        self.b0 = nn.Parameter(torch.randn(1))
        self.w = nn.Parameter(torch.randn(input_dim, output_dim-1))
        self.b = nn.Parameter(torch.randn(output_dim-1))
        self.f = torch.sin

        # Initialize parameters with Xavier/Glorot initialization
        nn.init.xavier_uniform_(self.w0)
        nn.init.xavier_uniform_(self.w)
        nn.init.zeros_(self.b0)
        nn.init.zeros_(self.b)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        v1 = torch.matmul(x, self.w0) + self.b0
        v2 = self.f(torch.matmul(x, self.w) + self.b)
        return torch.cat([v1, v2], -1)

class BitcoinFeeModel(nn.Module):
    """Enhanced Bitcoin Fee Prediction Model with fixed normalization"""
    def __init__(self, hist_input_dim: int, future_input_dim: int, config: ModelConfig):
        super(BitcoinFeeModel, self).__init__()
        self.time2vec = Time2Vec(1, config.time2vec_dim)
        
        # Multi-layer LSTM with dropout
        self.lstm = nn.LSTM(
            hist_input_dim + config.time2vec_dim,
            config.hidden_dim,
            num_layers=config.num_lstm_layers,
            dropout=config.dropout_rate,
            bidirectional=True,
            batch_first=True
        )
        
        # Multi-head attention mechanism
        self.attention = nn.MultiheadAttention(
            config.hidden_dim * 2,  # *2 for bidirectional
            num_heads=config.num_attention_heads,
            dropout=config.dropout_rate
        )
        
        lstm_output_dim = config.hidden_dim * 2  # *2 for bidirectional
        combined_dim = lstm_output_dim + future_input_dim
        
        # Deeper fully connected layers with layer normalization
        self.fc_layers = nn.Sequential(
            nn.Linear(combined_dim, config.hidden_dim),
            nn.LayerNorm(config.hidden_dim),
            nn.ReLU(),
            nn.Dropout(config.dropout_rate),
            nn.Linear(config.hidden_dim, config.hidden_dim // 2),
            nn.LayerNorm(config.hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(config.dropout_rate)
        )
        
        self.fc_out = nn.Linear(config.hidden_dim // 2, 1)
        self.num_predictions = config.max_prediction_length

    def forward(self, time_feature: torch.Tensor, x_hist: torch.Tensor, x_future: torch.Tensor) -> torch.Tensor:
        batch_size, seq_len, _ = x_hist.shape
        
        # Time encoding
        time_encoded = self.time2vec(time_feature).expand(batch_size, seq_len, -1)
        x_hist_time = torch.cat([x_hist, time_encoded], dim=-1)
        
        # LSTM processing
        lstm_out, (h_n, _) = self.lstm(x_hist_time)
        
        # Attention mechanism
        attn_out, _ = self.attention(
            lstm_out.transpose(0, 1),
            lstm_out.transpose(0, 1),
            lstm_out.transpose(0, 1)
        )
        attn_out = attn_out.transpose(0, 1)
        
        # Combine all LSTM layers and directions
        h_n = h_n.view(self.lstm.num_layers, 2, batch_size, -1)  # [num_layers, num_directions, batch, hidden]
        h_n = h_n[-1]  # Take the last layer [num_directions, batch, hidden]
        h_n = h_n.transpose(0, 1).reshape(batch_size, -1)  # [batch, num_directions * hidden]
        
        # Process each prediction step
        outputs = []
        for i in range(self.num_predictions):
            # Combine LSTM output with future features
            combined = torch.cat((h_n, x_future[:, i, :]), dim=1)
            hidden = self.fc_layers(combined)
            out = self.fc_out(hidden)
            outputs.append(out)
        
        return torch.cat(outputs, dim=1)

class Time2VecDataset(Dataset):
    """Enhanced dataset class with data augmentation"""
    def __init__(self, data: pd.DataFrame, historical_features: List[str],
                 future_known_features: List[str], target: str, config: ModelConfig,
                 augment: bool = True):
        self.data = data
        self.historical_features = historical_features
        self.future_known_features = future_known_features
        self.target = target
        self.config = config
        self.augment = augment

    def __len__(self) -> int:
        return len(self.data) - self.config.max_encoder_length - self.config.max_prediction_length + 1

    def augment_data(self, data: np.ndarray) -> np.ndarray:
        """Simple time series augmentation"""
        if not self.augment or random.random() > 0.5:
            return data
            
        # Random scaling
        scale_factor = np.random.normal(1.0, 0.1)
        data = data * scale_factor
        
        # Add small random noise
        noise = np.random.normal(0, 0.01, data.shape)
        data = data + noise
        
        return data

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
        encoder_start = idx
        encoder_end = idx + self.config.max_encoder_length
        decoder_end = encoder_end + self.config.max_prediction_length

        historical_data = self.data.iloc[encoder_start:encoder_end][self.historical_features]
        future_known_data = self.data.iloc[encoder_end:decoder_end][self.future_known_features]
        target_values = self.data.iloc[encoder_end:decoder_end][self.target]

        # Convert to numpy for augmentation
        historical_data = self.augment_data(historical_data.values)
        
        time_feature = torch.arange(0, self.config.max_encoder_length, dtype=torch.float32).unsqueeze(1)
        x_hist = torch.tensor(historical_data, dtype=torch.float)
        x_future = torch.tensor(future_known_data.values, dtype=torch.float)
        y = torch.tensor(target_values.values, dtype=torch.float)

        return time_feature, x_hist, x_future, y

@dataclass
class ModelMetrics:
    """Store model evaluation metrics"""
    mae: float
    rmse: float
    r2: float
    mape: float
    theil_u: float
    directional_accuracy: float

class BitcoinFeePredictor:
    """Enhanced Bitcoin Fee Predictor"""
    def __init__(self, config: ModelConfig = ModelConfig()):
        self.config = config
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        
        # Enhanced feature set
        self.historical_features = [
            'tx_count', 'mempool_size_mb', 'max_fee_rate', 'avg_fee_rate',
            'median_fee_rate', 'fee_rate_10th', 'fee_rate_90th', 'fee_rate_std',
            'mempool_min_fee', 'total_fee', 'mempool_usage', 'transaction_count',
            'block_weight', 'block_interval', 'hist_low_fee_ratio',
            'hist_med_fee_ratio', 'hist_high_fee_ratio', 'hist_fee_diversity',
            'hash_rate', 'bitcoin_price_usd'
        ]
        
        self.known_future_features = ['block_height', 'difficulty', 'block_version']
        self.target = 'block_median_fee_rate'
        
    def prepare_data(self, df: pd.DataFrame) -> pd.DataFrame:
        """Enhanced data preparation with feature engineering"""
        all_features = self.historical_features + self.known_future_features
        
        # Add time-based features
        df['hour'] = df['block_time'].dt.hour
        df['day_of_week'] = df['block_time'].dt.dayofweek
        df['is_weekend'] = df['block_time'].dt.dayofweek.isin([5, 6]).astype(int)
        
        # Log transform highly skewed features
        skewed_features = ['mempool_size_mb', 'max_fee_rate', 'total_fee', 'hash_rate']
        for feature in skewed_features:
            df[feature] = np.log1p(df[feature])
        
        # Add rolling statistics
        for window in [6, 12, 24]:
            df[f'fee_rate_ma_{window}'] = df['block_median_fee_rate'].rolling(window).mean()
            df[f'fee_rate_std_{window}'] = df['block_median_fee_rate'].rolling(window).std()
        
        # Add rate of change features
        df['fee_rate_pct_change'] = df['block_median_fee_rate'].pct_change()
        df['mempool_size_pct_change'] = df['mempool_size_mb'].pct_change()
        
        # Handle missing values
        df = df.fillna(method='ffill').fillna(method='bfill')
        
        # Robust scaling with clipping
        scaler = RobustScaler()
        df[all_features] = scaler.fit_transform(df[all_features])
        
        # Clip extreme values
        for col in all_features:
            df[col] = df[col].clip(-5, 5)
        
        return df

    def train_model(self, train_data: pd.DataFrame) -> BitcoinFeeModel:
        """Enhanced training process"""
        train_size = int(len(train_data) * 0.8)
        train_subset = train_data.iloc[:train_size]
        
        train_dataset = Time2VecDataset(train_subset, self.historical_features,
                                      self.known_future_features, self.target, self.config)
        train_loader = DataLoader(train_dataset, batch_size=self.config.batch_size,
                                shuffle=True, drop_last=True)
        val_loader = self.create_validation_dataloader(train_data)
        
        model = BitcoinFeeModel(len(self.historical_features),
                               len(self.known_future_features), self.config).to(self.device)
        
        # Combined loss function
        criterion = lambda y_pred, y_true: (
            0.7 * nn.HuberLoss()(y_pred, y_true) +
            0.3 * nn.L1Loss()(y_pred, y_true)
        )
        
        # Optimizer with weight decay
        optimizer = optim.AdamW(
            model.parameters(),
            lr=self.config.learning_rate,
            weight_decay=0.01,
            betas=(0.9, 0.999)
        )
        
        scheduler = optim.lr_scheduler.OneCycleLR(
    optimizer,
    max_lr=self.config.learning_rate,
    epochs=self.config.num_epochs,
    steps_per_epoch=len(train_loader),
    pct_start=0.1,
    anneal_strategy='cos',
    div_factor=1e3,     # Initial LR will be max_lr / div_factor
    final_div_factor=1e4  # Final LR will be max_lr / final_div_factor
)

        
        early_stopping = EarlyStopping(
            patience=self.config.patience,
            min_delta=self.config.min_delta
        )
        
        best_model_state = None
        best_val_loss = float('inf')
        
        for epoch in range(self.config.num_epochs):
            model.train()
            running_loss = 0.0
            
            for time_feature, x_hist, x_future_known, y in train_loader:
                time_feature = time_feature.to(self.device)
                x_hist = x_hist.to(self.device)
                x_future_known = x_future_known.to(self.device)
                y = y.to(self.device)

                outputs = model(time_feature, x_hist, x_future_known)
                loss = criterion(outputs, y)

                optimizer.zero_grad()
                loss.backward()
                
                # Gradient clipping
                torch.nn.utils.clip_grad_norm_(model.parameters(), self.config.clip_grad_norm)
                
                optimizer.step()
                scheduler.step()
                
                running_loss += loss.item()
            
            train_loss = running_loss / len(train_loader)
            val_loss = self.validate_model(model, val_loader, criterion)
            
            current_lr = optimizer.param_groups[0]['lr']
            print(f"Epoch {epoch+1}/{self.config.num_epochs}, "
                  f"Train Loss: {train_loss:.4f}, "
            f"Val Loss: {val_loss:.4f}, "
                  f"Learning Rate: {current_lr:.6f}")
            
            # Save best model
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                best_model_state = model.state_dict().copy()
            
            # Early stopping check
            if early_stopping(val_loss, model):
                print("Early stopping triggered")
                break
        
        # Load best model
        if best_model_state is not None:
            model.load_state_dict(best_model_state)
        
        return model

    def validate_model(self, model: nn.Module, val_loader: DataLoader, 
                      criterion: callable) -> float:
        """Enhanced validation process"""
        model.eval()
        val_loss = 0.0
        predictions = []
        actuals = []
        
        with torch.no_grad():
            for time_feature, x_hist, x_future_known, y in val_loader:
                time_feature = time_feature.to(self.device)
                x_hist = x_hist.to(self.device)
                x_future_known = x_future_known.to(self.device)
                y = y.to(self.device)

                outputs = model(time_feature, x_hist, x_future_known)
                loss = criterion(outputs, y)
                val_loss += loss.item()
                
                predictions.extend(outputs.cpu().numpy())
                actuals.extend(y.cpu().numpy())
        
        return val_loss / len(val_loader)

    def create_validation_dataloader(self, train_data: pd.DataFrame) -> DataLoader:
        """Create validation dataloader with enhanced settings"""
        val_size = int(len(train_data) * 0.2)
        val_data = train_data.iloc[-val_size:]
        val_dataset = Time2VecDataset(
            val_data, 
            self.historical_features,
            self.known_future_features, 
            self.target, 
            self.config,
            augment=False  # No augmentation for validation
        )
        return DataLoader(
            val_dataset, 
            batch_size=self.config.batch_size,
            shuffle=False, 
            drop_last=True
        )

    def predict(self, model: BitcoinFeeModel, data: pd.DataFrame) -> np.ndarray:
        """Enhanced prediction with uncertainty estimation"""
        model.eval()
        
        # Prepare historical data
        hist_data = data.tail(self.config.max_encoder_length)[self.historical_features].values
        
        # Prepare future known features
        future_known_data = np.zeros((self.config.max_prediction_length, len(self.known_future_features)))
        last_block_height = data['block_height'].iloc[-1]
        last_difficulty = data['difficulty'].iloc[-1]
        last_block_version = data['block_version'].iloc[-1]
        
        for i in range(self.config.max_prediction_length):
            future_known_data[i] = [last_block_height + i + 1, last_difficulty, last_block_version]
        
        # Convert to tensors
        time_feature = torch.arange(0, self.config.max_encoder_length, dtype=torch.float32).unsqueeze(1).to(self.device)
        x_hist = torch.tensor(hist_data, dtype=torch.float).unsqueeze(0).to(self.device)
        x_future_known = torch.tensor(future_known_data, dtype=torch.float).unsqueeze(0).to(self.device)

        predictions = []
        
        # Monte Carlo dropout for uncertainty estimation
        model.train()  # Enable dropout for MC estimation
        n_samples = 10
        
        with torch.no_grad():
            for _ in range(n_samples):
                pred = model(time_feature, x_hist, x_future_known).cpu().numpy().flatten()
                predictions.append(pred)
        
        # Calculate mean and std of predictions
        predictions = np.array(predictions)
        mean_predictions = np.mean(predictions, axis=0)
        std_predictions = np.std(predictions, axis=0)
        
        return mean_predictions, std_predictions

    @staticmethod
    def calculate_metrics(y_true: np.ndarray, y_pred: np.ndarray) -> ModelMetrics:
        """Enhanced metric calculation"""
        def mean_absolute_percentage_error(y_true, y_pred):
            return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

        def directional_accuracy(y_true, y_pred):
            return np.mean((np.diff(y_true) * np.diff(y_pred)) > 0) * 100

        def theil_u(y_true, y_pred):
            y_true_pct = (y_true[1:] - y_true[:-1]) / y_true[:-1]
            y_pred_pct = (y_pred[1:] - y_true[:-1]) / y_true[:-1]
            numerator = np.sqrt(np.mean((y_true_pct - y_pred_pct)**2))
            denominator = np.sqrt(np.mean(y_true_pct**2)) + np.sqrt(np.mean(y_pred_pct**2))
            return numerator / denominator if denominator != 0 else float('inf')

        return ModelMetrics(
            mae=mean_absolute_error(y_true, y_pred),
            rmse=np.sqrt(mean_squared_error(y_true, y_pred)),
            r2=r2_score(y_true, y_pred),
            mape=mean_absolute_percentage_error(y_true, y_pred),
            theil_u=theil_u(y_true, y_pred),
            directional_accuracy=directional_accuracy(y_true, y_pred)
        )

    def plot_predictions(self, y_true: np.ndarray, y_pred: np.ndarray, 
                        y_std: np.ndarray = None, title: str = "Bitcoin Fee Predictions"):
        """Enhanced visualization with uncertainty"""
        plt.figure(figsize=(15, 8))
        
        # Plot actual values
        plt.plot(range(len(y_true)), y_true, 
                label='Actual', alpha=0.3)
        
        # Plot predicted values
        plt.plot(range(len(y_pred)), y_pred, 
                label='Predicted', alpha=0.3)
        
        # Add confidence bands if standard deviation is provided
        # if y_std is not None:
        #     plt.fill_between(
        #         range(len(y_pred)),
        #         y_pred - 2*y_std,
        #         y_pred + 2*y_std,
        #         color='gray',
        #         alpha=0.2,
        #         label='95% Confidence Interval'
        #     )
        # 
        plt.title(title)
        plt.xlabel('Block Number')
        plt.ylabel('Fee Rate')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        return plt.gcf()

    def cross_validate(self, df: pd.DataFrame, n_splits: int = 5) -> Tuple[List[ModelMetrics], List[np.ndarray], List[np.ndarray]]:
        """Enhanced cross-validation with uncertainty estimation"""
        df = self.prepare_data(df)
        n_samples = len(df)
        metrics_list = []
        all_y_true = []
        all_y_pred = []
        all_y_std = []
        
        for fold_idx in range(n_splits):
            train_end = n_samples - (n_splits - fold_idx) * self.config.max_prediction_length
            test_start = train_end
            test_end = test_start + self.config.max_prediction_length
            
            # Split data
            train_data = df.iloc[:train_end]
            test_data = df.iloc[test_start:test_end]
            
            # Train model
            model = self.train_model(train_data)
            
            # Make predictions
            predictions, std_predictions = self.predict(model, train_data)
            
            # Get actual values
            y_true = test_data[self.target].values
            
            # Calculate metrics
            metrics = self.calculate_metrics(y_true, predictions)
            metrics_list.append(metrics)
            
            all_y_true.extend(y_true)
            all_y_pred.extend(predictions)
            all_y_std.extend(std_predictions)
            
            # Plot fold results
            self.plot_predictions(
                y_true, 
                predictions, 
                std_predictions,
                f'Actual vs Predicted Bitcoin Fee Rates (Fold {fold_idx + 1})'
            )
            plt.show()
            
        return metrics_list, all_y_true, all_y_pred, all_y_std

def main():
    # Set random seeds for reproducibility
    random.seed(42)
    np.random.seed(42)
    torch.manual_seed(42)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(42)
        torch.cuda.manual_seed_all(42)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    
    # Load and preprocess data
    df = pd.read_csv('real_time_cleaned.csv')
    df['block_time'] = pd.to_datetime(df['block_time'])
    df = df.sort_values('block_time')
    
    # Initialize predictor and perform cross-validation
    start_time = time.time()
    predictor = BitcoinFeePredictor()
    metrics_list, all_y_true, all_y_pred, all_y_std = predictor.cross_validate(df)
    end_time = time.time()
    
    # Calculate and print average metrics
    avg_metrics = ModelMetrics(
        mae=np.mean([m.mae for m in metrics_list]),
        rmse=np.mean([m.rmse for m in metrics_list]),
        r2=np.mean([m.r2 for m in metrics_list]),
        mape=np.mean([m.mape for m in metrics_list]),
        theil_u=np.mean([m.theil_u for m in metrics_list]),
        directional_accuracy=np.mean([m.directional_accuracy for m in metrics_list])
    )
    
    print("\nCross-Validation Performance:")
    print(f"Mean Absolute Error (CV): {avg_metrics.mae:.4f}")
    print(f"Root Mean Squared Error (CV): {avg_metrics.rmse:.4f}")
    print(f"R-squared Score (CV): {avg_metrics.r2:.4f}")
    print(f"MAPE (CV): {avg_metrics.mape:.4f}%")
    print(f"Theil's U Statistic (CV): {avg_metrics.theil_u:.4f}")
    print(f"Directional Accuracy (CV): {avg_metrics.directional_accuracy:.4f}%")
    print(f"Training and Prediction time: {end_time - start_time:.2f}s")
    
    # Plot final results
    predictor.plot_predictions(
        np.array(all_y_true),
        np.array(all_y_pred),
        np.array(all_y_std),
        'Actual vs Predicted Bitcoin Fee Rates (All Folds Combined)'
    )
    plt.show()

if __name__ == "__main__":
    main()
```

```python
def train_and_evaluate_final_model():
    try:
        # Set random seeds
        random.seed(42)
        np.random.seed(42)
        torch.manual_seed(42)
        if torch.cuda.is_available():
            torch.cuda.manual_seed(42)
            torch.cuda.manual_seed_all(42)
            torch.backends.cudnn.deterministic = True
            torch.backends.cudnn.benchmark = False

        # Load training and test data
        df = pd.read_csv('real_time_cleaned.csv')
        df['block_time'] = pd.to_datetime(df['block_time'])
        df = df.sort_values('block_time')

        test_df = pd.read_csv('test.csv')
        actual_values = test_df['block_median_fee_rate']

        # Initialize predictor
        predictor = BitcoinFeePredictor()
        
        # Prepare data
        print("Preparing data...")
        df_prepared = predictor.prepare_data(df)
        
        # Train model on full dataset
        print("\nTraining model on full dataset...")
        model = predictor.train_model(df_prepared)
        
        # Generate predictions
        print("\nGenerating predictions...")
        predictions, std_predictions = predictor.predict(model, df_prepared)
        
        # Calculate metrics
        metrics = predictor.calculate_metrics(actual_values.values, predictions)
        
        # Print performance metrics
        print("\nFinal Model Performance:")
        print(f"Mean Absolute Error (MAE): {metrics.mae:.4f}")
        print(f"Root Mean Squared Error (RMSE): {metrics.rmse:.4f}")
        print(f"Theil's U Statistic: {metrics.theil_u:.4f}")
        
        # Create plot
        plt.figure(figsize=(15, 8))

        # Plot actual values
        plt.plot(actual_values.index, actual_values, 
                 label='Actual Fee Rate', linewidth=2)

        # Plot predicted values
        plt.plot(actual_values.index, predictions[:len(actual_values)], 
                 label='Predicted Fee Rate', linestyle='--', linewidth=2)



        # Add title and labels
        plt.title('Actual vs Predicted Bitcoin Fee Rate', fontsize=16)
        plt.xlabel('Time', fontsize=14)
        plt.ylabel('Fee Rate (sat/vB)', fontsize=14)

        # Add legend and grid
        plt.legend(fontsize=12)
        plt.grid(alpha=0.3)

        # Rotate x-axis labels
        plt.xticks(rotation=45)

        # Adjust layout
        plt.tight_layout()
        plt.show()
        
        # Save results
        results_df = pd.DataFrame({
            'block_time': test_df['block_time'],
            'actual_fee_rate': actual_values,
            'predicted_fee_rate': predictions[:len(actual_values)],
            'prediction_std': std_predictions[:len(actual_values)]
        })
        
        results_df.to_csv('time2vec_final_predictions.csv', index=False)
        print("\nResults saved to 'time2vec_final_predictions.csv'")
        
        return results_df, metrics

    except Exception as e:
        print(f"Error in final evaluation process: {str(e)}")
        raise

if __name__ == "__main__":
    results_df, final_metrics = train_and_evaluate_final_model()
```
